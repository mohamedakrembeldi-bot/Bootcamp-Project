```python
import os
import numpy as np
import cv2
from google.colab import drive
from tensorflow.keras.models import load_model

# 1. Mounting the Drive
drive.mount('/content/drive')

# 2. Define the Bridge paths
PROJECT_PATH = '/content/drive/MyDrive/Face_Recognition_Project/'
DATA_PATH = os.path.join(PROJECT_PATH, 'processed_faces/')

from keras_facenet import FaceNet

# Initialize the model
embedder = FaceNet()

print("вњ… FaceNet model loaded.")
```

```python
def get_embedding(model, face_pixels):
    # Scale pixel (standardization)
    face_pixels = face_pixels.astype('float32')
    # FaceNet expects a 4D array (1, 160, 160, 3)
    samples = np.expand_dims(face_pixels, axis=0)
    # Get the embedding (128-dimensional vector)
    yhat = model.embeddings(samples)
    return yhat[0]

def load_dataset(directory):
    X, y = [], []
    # Loop through each person's folder
    for subdir in os.listdir(directory):
        path = os.path.join(directory, subdir)
        if not os.path.isdir(path):
            continue

        print(f"Loading embeddings for: {subdir}")
        for name in os.listdir(path):
            img_path = os.path.join(path, name)
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Get the mathematical signature
            embedding = get_embedding(embedder, img)
            X.append(embedding)
            y.append(subdir) # The folder name is our label

    return np.asarray(X), np.asarray(y)

# Execute the extraction
trainX, trainy = load_dataset(DATA_PATH)
print(f"вњ… Extracted {trainX.shape[0]} embeddings for {len(np.unique(trainy))} people.")

# Save the extraction to drive
np.savez_compressed(os.path.join(PROJECT_PATH, 'celebrity-embeddings.npz'), trainX, trainy)
print("вњ… Embeddings saved to Drive!")
```
