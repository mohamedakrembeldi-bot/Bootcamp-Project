```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import joblib
import os
from google.colab import drive

# 1. Mount Drive
drive.mount('/content/drive')

# 2. Path to extacted features
PROJECT_PATH = '/content/drive/MyDrive/Face_Recognition_Project/'
data = np.load(os.path.join(PROJECT_PATH, 'celebrity-embeddings.npz'))

trainX, trainy = data['arr_0'], data['arr_1']
print(f"✅ Loaded {trainX.shape[0]} samples.")
```

```python
# Encoding
out_encoder = LabelEncoder()
out_encoder.fit(trainy)
trainy_encoded = out_encoder.transform(trainy)

# Split data
X_train, X_test, y_train, y_test = train_test_split(trainX, trainy_encoded, test_size=0.2, random_state=42)

print("✅ Labels encoded and data split.")
```

```python
# Initialize the SVM with 'probability=True' so we can see how confident it is
model = SVC(kernel='linear', probability=True)
model.fit(X_train, y_train)

# Predict on the test set to see how we did
yhat_test = model.predict(X_test)
score_test = accuracy_score(y_test, yhat_test)

print(f"✅ Training Complete!")
print(f"🔥 Accuracy: {score_test * 100:.2f}%")
```

```python
# Save the model and the encoder to Drive
joblib.dump(model, os.path.join(PROJECT_PATH, 'face_recognizer_model.pkl'))
joblib.dump(out_encoder, os.path.join(PROJECT_PATH, 'label_encoder.pkl'))

print(" SVM model saved to Drive")
```
